```python
# while1.py
a = 6
while a > 0:
    print(a)
    a -= 1 
print("離開後 a 為", a)
```

```python
## 計算2-10 所有偶數的總和
```

```python
total=0 #總和開始 是0
i=2
while(i<=10): # 有括號連續執行，沒有括號要控一格
    total +=i
    print(f"第{i//2}次迴圈的i={i},總合為{total}") #大括號裡面運算式
    i+=2
```

```python
even_total = 0
odd_total = 0

for i in range(1,101):
    if i % 2 == 0:
        even_total += i
    else:
        odd_total += i
print(f"1加到100偶數的加總為{even_total}")
print(f"1加到100奇數的加總為{odd_total}")
    
```
