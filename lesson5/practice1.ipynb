```python
#巢狀判斷
furry = False #已儲存布林值
small= False
if furry: 
    print("這是有毛的")
    if small: 
        print("這是小的")
        print("牠是貓")
    else: 
        print("這是大的")
        print("牠是熊")
        
else: 
    print("代表這是沒有毛")
    if small:
        print("這是小的")
        print("這是小蜥蜴")
    else: 
        print("這是大的")
        print("他是人類或眉毛的熊")
```

```python
x=int(input("請輸入x:"))# 要求使用者輸入  裡面要放字串  改為數字要用int 但會是整數
y=int(input("請輸入y:"))

if x>=0:
    root=x**(1/y)
else:
    if y%2==0: 
        root="虛數"
    else:
        root=-(-x)**(1/y)
print("root=",root)
```
